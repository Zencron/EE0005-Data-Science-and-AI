```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
houseData = pd.read_csv('train.csv')
houseData.head()
```

```python
houseNumData = pd.DataFrame(houseData[['LotArea', 'GrLivArea', 'TotalBsmtSF', 'GarageArea', 'SalePrice']])
houseNumData.head()
```

```python
houseNumData.describe()
```

```python
f, axes = plt.subplots(1,5, figsize=(24, 4))
sb.distplot(houseNumData['LotArea'] , ax = axes[0])
sb.distplot(houseNumData['GrLivArea'] , ax = axes[1])
sb.distplot(houseNumData['TotalBsmtSF'] , ax = axes[2])
sb.distplot(houseNumData['GarageArea'] , ax = axes[3])
sb.distplot(houseNumData['SalePrice'] , ax = axes[4])
```

```python
houseNumData.corr()
```

```python
sb.heatmap(houseNumData.corr() , annot = True)
```
