```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
houseData = pd.read_csv('train.csv')
salePrice = pd.DataFrame(houseData['SalePrice'])
lotArea = pd.DataFrame(houseData['LotArea'])
totalBsmtSF = pd.DataFrame(houseData['TotalBsmtSF'])
garageArea = pd.DataFrame(houseData['GarageArea'])
```

```python
salePriceTrain , salePriceTest , lotAreaTrain , lotAreaTest , totalBsmtSFTrain , totalBsmtSFTest , garageAreaTrain , garageAreaTest = train_test_split(salePrice , lotArea , totalBsmtSF , garageArea , train_size = 0.75)
```

```python
lotAreaLinReg = LinearRegression()
totalBsmtSFLinReg = LinearRegression()
garageAreaLinReg = LinearRegression()
```

```python
lotAreaLinReg.fit(lotAreaTrain , salePriceTrain)
totalBsmtSFLinReg.fit(totalBsmtSFTrain , salePriceTrain)
garageAreaLinReg.fit(garageAreaTrain , salePriceTrain)
```

```python
lotAreaTestPredict = lotAreaLinReg.predict(lotAreaTest)
totalBsmtSFTestPredict = totalBsmtSFLinReg.predict(totalBsmtSFTest)
garageAreaTestPredict = garageAreaLinReg.predict(garageAreaTest)

f , axes = plt.subplots( 1 , 3 , figsize = (24 , 12))
axes[0].scatter(lotAreaTest , salePriceTest)
axes[0].scatter(lotAreaTest , lotAreaTestPredict)
axes[1].scatter(totalBsmtSFTest , salePriceTest)
axes[1].scatter(totalBsmtSFTest , totalBsmtSFTestPredict)
axes[2].scatter(garageAreaTest , salePriceTest)
axes[2].scatter(garageAreaTest , garageAreaTestPredict)
```

```python
print("Goodness of Fit of Model \tLotArea")
print("Explained Variance (R^2) \t:", lotAreaLinReg.score(lotAreaTest, salePriceTest))
print("Mean Squared Error (MSE) \t:", mean_squared_error(salePriceTest, lotAreaTestPredict))

print("\nGoodness of Fit of Model \t TotalBsmtSF")
print("Explained Variance (R^2) \t:", totalBsmtSFLinReg.score(totalBsmtSFTest, salePriceTest))
print("Mean Squared Error (MSE) \t:", mean_squared_error(salePriceTest, totalBsmtSFTestPredict))

print("\nGoodness of Fit of Model \tGarageArea")
print("Explained Variance (R^2) \t:", garageAreaLinReg.score(garageAreaTest, salePriceTest))
print("Mean Squared Error (MSE) \t:", mean_squared_error(salePriceTest, garageAreaTestPredict))
```
